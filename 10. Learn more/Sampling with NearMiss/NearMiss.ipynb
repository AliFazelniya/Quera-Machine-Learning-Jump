نمونه‌کاهی با NearMiss
</font>
</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def plot(X, y, title):
    plt.figure(figsize=(10,7))
    counter = Counter(y)
    for label, num in counter.items():
        row_ix = np.where(y == label)[0]
        plt.scatter(X[row_ix, 0], X[row_ix, 1], label='class='+str(label)+' ['+str(num)+' samples]', s=64)
    plt.legend()
    plt.title(title)
    plt.xlim([-1.7, 1.7])
    plt.ylim([-3.5, 2.5])
    plt.show()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
 مجموعه‌داده
</font>
</h2>

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=20, n_features=2, n_redundant=0, weights=[0.75], flip_y=0, random_state=1)
plot(X, y, 'Original Dataset')
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
تکنیک نمونه‌کاهی تصادفی
</font>
</h2>

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(sampling_strategy='majority', random_state=42)
X_ru, y_ru = undersample.fit_resample(X, y)
plot(X_ru, y_ru , 'Random Undersampling')
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
تکنیک NearMiss-1
</font>
</h2>

```python
from imblearn.under_sampling import NearMiss

undersample = NearMiss(version=1, n_neighbors=3)
X_1, y_1 = undersample.fit_resample(X, y)
plot(X_1, y_1, 'NearMiss-1')
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
تکنیک NearMiss-2
</font>
</h2>

```python
from imblearn.under_sampling import NearMiss

undersample = NearMiss(version=2, n_neighbors=3)
X_2, y_2 = undersample.fit_resample(X, y)
plot(X_2, y_2, 'NearMiss-2')
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
تکنیک NearMiss-3
</font>
</h2>

```python
from imblearn.under_sampling import NearMiss

undersample = NearMiss(version=3, n_neighbors_ver3=3)
X_3, y_3 = undersample.fit_resample(X, y)
plot(X_3, y_3, 'NearMiss-3')
```
