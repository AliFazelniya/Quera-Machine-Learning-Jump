```python
preprocessor = Preprocessor(df)
X_processed = preprocessor.transform()

print('Processed data shape:', X_processed.shape)
print('Columns:', X_processed.columns.tolist())
print()
print(X_processed.describe())
X_processed.head()
```

```python
X = X_processed.values

inertias = []
sil_scores = []
k_values = range(2, 12)

for k in k_values:
    km = KMeans(n_clusters=k, n_init=50, max_iter=500, random_state=42)
    labels = km.fit_predict(X)
    inertias.append(km.inertia_)
    sil = silhouette_score(X, labels)
    sil_scores.append(sil)
    grader_score = 100 * (sil + 1) / 2
    print(f'k={k:2d}: inertia={km.inertia_:8.2f}, silhouette={sil:.4f}, grader_score={grader_score:.2f}')

best_k = k_values[np.argmax(sil_scores)]
best_sil = max(sil_scores)
print(f'\n>>> Best k = {best_k} with silhouette = {best_sil:.4f} (grader score = {100*(best_sil+1)/2:.2f})')
```

```python
from sklearn.cluster import KMeans

model = KMeans(
    n_clusters=best_k,
    init='k-means++',  
    n_init=50,        
    max_iter=500,      
    random_state=42,
)

model.fit(X)
labels = model.labels_

print(f'KMeans fitted with k={best_k}')
print('Cluster distribution:')
print(pd.Series(labels).value_counts().sort_index())
```
