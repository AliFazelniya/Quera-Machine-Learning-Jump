```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(200) * 4 - 2
y = 3 * x**2 + 2*x + 5 + np.random.uniform(-2, 2, 200)

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
رگرسیون خطی
</font>
</h2>

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=42)

lr = LinearRegression()
lr.fit(x_train.reshape(-1, 1), y_train)
y_pred = lr.predict(x_test.reshape(-1, 1))
print(r2_score(y_test, y_pred))
```

```python
plt.plot(x, lr.predict(x.reshape(-1, 1)), "r", linewidth = 3)
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
رگرسیون چندجمله‌ای
</font>
</h2>

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=True)
x_train_trans = poly.fit_transform(x_train.reshape(-1 ,1))
x_test_trans = poly.transform(x_test.reshape(-1, 1))

lr = LinearRegression()
lr.fit(x_train_trans, y_train)
y_pred = lr.predict(x_test_trans)
print(r2_score(y_test, y_pred))
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
x_new = poly.transform(np.sort(x).reshape(-1, 1))
plt.plot(np.sort(x), lr.predict(x_new), "r", linewidth = 3)
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
