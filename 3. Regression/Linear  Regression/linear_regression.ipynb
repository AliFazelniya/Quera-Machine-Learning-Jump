```python
import pandas as pd

df = pd.DataFrame()
df['area'] = [130, 140, 69, 180, 175, 100, 220]
df['price'] = [6.5, 9.6, 2.5, 13, 17.5, 7.4, 16]
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
آموزش مدل
</font>
</h2>

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(df.area.to_numpy().reshape(-1,1), df.price)
```

```python
print(model.coef_)
print(model.intercept_)
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
مصورسازی نتیجه یادگیری
</font>
</h2>


```python
import numpy as np
import matplotlib.pyplot as plt

a = model.intercept_
b = model.coef_[0]

x = np.linspace(60, 220, 100)
y = a + (x*b)

fig, ax = plt.subplots()
ax.scatter(df.area, df.price, c='blue')
ax.plot(x,y, c='green')

ax.set_xlabel('area(m2)')
ax.set_ylabel('price(billion tomans)')
```

```python
test = pd.DataFrame()
test['area'] = [80, 110, 230]
test['price'] = [5, 5.5, 17]
y_pred = model.predict(test.area.to_numpy().reshape(-1,1))
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(test.price, y_pred))
```
